```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('features_df.pkl')
```

```python
corr_hp_df = df[['CCMM_adopted_percent', 'non_cog', 'innov', 'percep', 'cons', 'CCMM_eff_mean', 'soc', 'agroconcept', 
    'age', 'educ', 'total_agr_land', 'farmtype', 'organic', 'share', 'tot_freq_qual']]
```

```python
plt.figure(figsize=(16, 6))

mask = np.triu(np.ones_like(corr_hp_df.corr(), dtype=np.bool))
heatmap = sns.heatmap(corr_hp_df.corr(), mask=mask, vmin=-1, vmax=1, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```
