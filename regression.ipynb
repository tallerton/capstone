```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_pickle('features_df.pkl')
```

```python
reg1 = smf.ols(
    formula = \
    'CCMM_adopted_percent ~ weak_ties', 
    data = df
).fit()

ols_robust = reg1.get_robustcov_results(cov_type= 'HC1')

print(ols_robust.summary())

reg1 = smf.ols(
    formula = \
    'CCMM_adopted_percent ~ weak_ties + non_cog + innov + percep + cons + CCMM_eff_mean + soc +\
    agroconcept + age + educ + total_agr_land + farmtype + organic + share', 
    data = df
).fit()

ols_robust = reg1.get_robustcov_results(cov_type= 'HC1')

print(ols_robust.summary())
```
